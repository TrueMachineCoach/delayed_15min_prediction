Predict whether a flight will be delayed for more than 15 minutes.

Features:

    Month, DayofMonth, DayOfWeek
    DepTime – departure time
    UniqueCarrier – code of a company-career
    Origin – flight origin
    Dest – flight destination
    Distance, distance between Origin and Dest airports
    dep_delayed_15min – target

```python
#necessory analysis tools importing
#creating a dataframe's object from csv file
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns

df = pd.read_csv('flight_delays_train.csv')
```

```python
df.head(n=10)
```

```python
#7 categorical features
#2 number features
df.info()
```

```python
#LabelEncoding categorical features of current dataset by numerical values
from sklearn.preprocessing import LabelEncoder
LE = LabelEncoder()
df['Month'] = LE.fit_transform(df['Month']) + 1
df['DayofMonth'] = LE.fit_transform(df['DayofMonth']) + 1
df['DayOfWeek'] = LE.fit_transform(df['DayOfWeek']) + 1
df['UniqueCarrier'] = LE.fit_transform(df['UniqueCarrier']) + 1
```

```python
#  visual interpretance: fraction of delayed flights over all dataset per month
plt.bar(df[df['dep_delayed_15min']=='Y']['Month'].
        value_counts().sort_index().index,
        df[df['dep_delayed_15min']=='Y']['Month'].value_counts().sort_index()/
        df['Month'].value_counts().sort_index() * 100,
        width=0.4, color='blue')
plt.plot(df[df['dep_delayed_15min']=='Y']['Month'].
        value_counts().sort_index().index,
        df[df['dep_delayed_15min']=='Y']['Month'].value_counts().sort_index()/
        df['Month'].value_counts().sort_index() * 100,
        color='brown', linewidth=3.7)
plt.xticks(range(1, 13, 1))
plt.xlabel('month number', fontsize=12, labelpad=10)
plt.ylabel('delayed flight\'s fraction, %', fontsize=12, labelpad=10)
plt.grid(True)
plt.tight_layout()
```

```python
#  visual interpretance: fraction of delayed flights over all dataset per DayofWeek
plt.figure(figsize=(6, 4))
plt.bar(df[df['dep_delayed_15min']=='Y']['DayOfWeek'].
        value_counts().sort_index().index,
        df[df['dep_delayed_15min']=='Y']['DayOfWeek'].value_counts().sort_index()/
        df['DayOfWeek'].value_counts().sort_index() * 100,
        width=0.4, color='pink')
plt.plot(df[df['dep_delayed_15min']=='Y']['DayOfWeek'].
        value_counts().sort_index().index,
        df[df['dep_delayed_15min']=='Y']['DayOfWeek'].value_counts().sort_index()/
        df['DayOfWeek'].value_counts().sort_index() * 100,
        color='brown', linewidth=3.7)
plt.xticks(range(1, 8, 1))
plt.xlabel('DayOfWeek', fontsize=12, labelpad=10)
plt.ylabel('delayed flight\'s fraction, %', fontsize=12, labelpad=10)
plt.grid(True)
plt.tight_layout()
```

```python
#Creating and filling a new subset to visualize distribution
#    of delayed flights fraction per DepTime(hour)
df.replace({'Y': 1, 'N': 0}, inplace=True)
ratio_delayed = []

for hour in range (0, 2400, 100):
    flights_per_hour = df[(df['DepTime'] > hour)
            & (df['DepTime'] <= hour + 100)]
    delayed = flights_per_hour[flights_per_hour['dep_delayed_15min']==1].shape[0]
    ratio_delayed.append((delayed/flights_per_hour.shape[0])*100)
```

```python
plt.figure(figsize=(10, 5))
plt.bar(range(0, 24, 1), ratio_delayed)
plt.plot(range(0, 24, 1), ratio_delayed, color='red')
plt.xticks(range(0, 24, 1))
plt.xlabel('hour\'s number', fontsize=12, labelpad=10)
plt.ylabel('delayed flight\'s fraction, %', fontsize=12, labelpad=10)
plt.grid(True)
```

```python
#conjuncting of features Origin and Dest by label encoder
df['Flight'] = LE.fit_transform(df['Origin'] + df['Dest'])

#filter and split the data into train and test
df = df[(df['DepTime'] >= 0) & (df['DepTime'] < 2400)]
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = \
train_test_split(df.drop(labels='dep_delayed_15min', axis=1),
                df['dep_delayed_15min'],
                test_size=0.3)
```

```python
from sklearn.preprocessing import OneHotEncoder

OHE = OneHotEncoder()
#extract a categorical features for future label encoding
X_train_categorical = X_train[['Month', 'DayofMonth', 'DayOfWeek', 'UniqueCarrier']]
X_test_categorical = X_test[['Month', 'DayofMonth', 'DayOfWeek', 'UniqueCarrier']]

#transforming train and test datasets by label encoding
X_train_OHE = OHE.fit_transform(X_train_categorical)
X_test_OHE = OHE.fit_transform(X_test_categorical)

#transforming datasets to pandas dataframe
X_test_OHE_df = pd.DataFrame(X_test_OHE.toarray(),
        columns=[i for i in range (1, 73)], index=X_test_categorical.index)
X_train_OHE_df = pd.DataFrame(X_train_OHE.toarray(),
        columns=[i for i in range (1, 73)], index=X_train_categorical.index)
```

```python
X_test_OHE_df
```

```python
X_train_OHE_df
```

```python
#addicting an numerical initial features
X_train_OHE_df['DepTime'] = X_train['DepTime']
X_train_OHE_df['Distance'] = X_train['Distance']
X_train_OHE_df['Flight'] = X_train['Flight']
X_test_OHE_df['DepTime'] = X_test['DepTime']
X_test_OHE_df['Distance'] = X_test['Distance']
X_test_OHE_df['Flight'] = X_test['Flight']
```

```python
#Scaling for Logistic Regression
from sklearn.preprocessing import StandardScaler

s_scale = StandardScaler()
X_train_OHE_df_scaled = s_scale.fit_transform(X_train_OHE_df)
X_test_OHE_df_scaled = s_scale.fit_transform(X_test_OHE_df)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import StratifiedKFold
#Cross-validation class importing and searching for optimal regularization parameter value
# Cross-validation folds creating by StratifiedKFold class,
# which allows to keep the same distribution character in each fold

logit_cv = GridSearchCV(LogisticRegression(n_jobs=-1),
                    param_grid={'C':np.logspace(-4, 4, 500)},
                    n_jobs=-1, scoring='roc_auc', cv=StratifiedKFold(n_splits=5))
```

```python
logit_cv.fit(X_train_OHE_df_scaled, Y_train)
```

```python
logit_cv.best_params_, logit_cv.best_score_
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(Y_test, logit_cv.predict(X_test_OHE_df_scaled))
```

```python
plt.figure(figsize=(10, 5))
plt.scatter(X_train['DepTime'], X_train['Flight'], c=Y_train)
plt.tight_layout()
```

```python
plt.figure(figsize=(10, 5))
plt.scatter(X_train['DepTime'], X_train['Flight'], c=logit_cv.predict(X_train_OHE_df_scaled))
plt.tight_layout()
```

```python
#xgboost is a good choice such as it makes possible to regularizate base trees
import xgboost

#finding for optimal number of a base trees with max_depth=3 by scoring roc_auc
boost_search_CV = GridSearchCV(xgboost.XGBClassifier(seed=17, n_jobs=-1, max_depth=3),
        param_grid={'n_estimators':range(200, 800, 100)},
        n_jobs=-1, scoring='roc_auc', cv=StratifiedKFold(n_splits=5))
```

```python
#dropping remaining garbge features
X_train.drop(labels=['Origin', 'Dest'], axis=1, inplace=True)
X_test.drop(labels=['Origin', 'Dest'], axis=1, inplace=True)
```

```python
boost_search_CV.fit(X_train, Y_train)
```

```python
boost_search_CV.best_params_, boost_search_CV.best_score_
```

```python
#Visualise 2-dimensional istribution again
plt.figure(figsize=(10, 5))
plt.scatter(X_train['DepTime'], X_train['Flight'], c=Y_train)
plt.xticks(range(0, 2500, 100))
plt.grid(b=True)
plt.tight_layout()
```

```python
plt.figure(figsize=(10, 5))
plt.scatter(X_train['DepTime'], X_train['Flight'],
            c=boost_search_CV.predict(X_train))
plt.xticks(range(0, 2500, 100))
plt.grid(b=True)
plt.tight_layout()
```

```python
#Coefficients for linear combination were chosen manually
predicted = 0.8*boost_search_CV.best_estimator_.predict_proba(X_train) +\
    0.2*logit_cv.best_estimator_.predict_proba(X_train_OHE_df_scaled)
```

```python
from sklearn.metrics import roc_auc_score
```

```python
#creating a class for atomate predicting
def scaling_and_onehotencoding(df):
    df['Flight'] = LE.fit_transform(df['Origin'] + df['Dest'])
    df = df[(df['DepTime'] >= 0) & (df['DepTime'] < 2400)]
    Y = df['dep_delayed_15min'].replace({'N':0, 'Y':1})
    df.drop(labels=['Origin', 'Dest', 'dep_delayed_15min'], axis=1, inplace=True)
#    OHE = OneHotEncoder()
    df_categorical = df[['Month', 'DayofMonth', 'DayOfWeek', 'UniqueCarrier']]
    df['Month'] = LE.fit_transform(df['Month'])
    df['DayofMonth'] = LE.fit_transform(df['DayofMonth'])
    df['DayOfWeek'] = LE.fit_transform(df['DayOfWeek'])
    df['UniqueCarrier'] = LE.fit_transform(df['UniqueCarrier'])
    df_OHE = OHE.fit_transform(df_categorical)
    df_OHE_df = pd.DataFrame(df_OHE.toarray(),
            columns=[i for i in range (1, df_OHE.toarray().shape[1] + 1)],
                             index=df_categorical.index)
    df_OHE_df['DepTime'] = df['DepTime']
    df_OHE_df['Distance'] = df['Distance']
    df_OHE_df['Flight'] = df['Flight']
#    s_scale = StandardScaler()
    df_OHE_df_scaled = s_scale.fit_transform(df_OHE_df)
    return df_OHE_df_scaled, df, Y
```

```python
gh, jk, Y = scaling_and_onehotencoding(df.copy())
```
